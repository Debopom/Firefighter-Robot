```python
def getBboxes(pred) : 
    predictions = pred['predictions']
    image_path = predictions[0]['image_path']  # Assuming it's the same for all predictions
    image = cv2.cvtColor(image_path, cv2.COLOR_BGR2RGB)

    # Iterate through predictions and draw bounding boxes
    for prediction in predictions:
        x, y, w, h = int(prediction['x']), int(prediction['y']), int(prediction['width']), int(prediction['height'])
        label = prediction['class']
        confidence = prediction['confidence']

        # Draw bounding box on the image
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Add label and confidence score to the bounding box
        text = f"{label}: {confidence:.2f}"
        cv2.putText(image, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return image
```
